```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
year = np.array([2018, 2019, 2020, 2021, 2022, 2023, 2024, 2025])
revenue = np.array([4.8, 5.3, 2.1, 3.4, 5.9, 6.5, 7.2, 8.1])

plt.figure()
plt.plot(year, revenue)

plt.title('Action Genre Box Office Revenue (2018-2025)')
plt.xlabel('Year')
plt.ylabel('Revenue (billions USD)')

plt.show()
```

```python
coffee = np.array([0, 1, 2, 3, 4, 5])
energy = np.array([3, 6, 8, 9, 8.5, 7])

plt.plot(coffee, energy)
```

```python
years=np.array([1980, 1990, 2000, 2010, 2020])
house_prices=np.array([64600, 123000, 169000, 221800, 336900])
plt.figure()
plt.plot(years, house_prices)
plt.title('Average House Prices in the US (1980-2020)')
```

```python
time = np.array([10, 20, 30, 40, 50, 60])

uber = np.array([59.99, 82.43, 66.34, 62.15, 60.99, 55.21])
lyft = np.array([65.81, 76.11, 72.45, 59.22, 55.26, 57.23])

plt.figure()

plt.plot(time, uber,label='Uber')
plt.plot(time, lyft,label='Lyft')
plt.title('Average Ride Cost Over Time')
plt.xlabel('Time (minutes)')    
plt.ylabel('Cost (USD)')
plt.legend()
plt.show()
```

```python
days = np.array(['Mon', 'Tue', 'Wed', 'Thu', 'Fri'])
emails = np.array([45, 38, 70, 49, 30])

plt.figure()
plt.plot(days, emails, color='purple', linestyle='-.', marker='o')

plt.title('Emails received on the week of 12/2')
plt.xlabel('Day of the Week')
plt.ylabel('Number of Emails')

plt.show()
```
