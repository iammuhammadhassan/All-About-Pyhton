```python
def write_song_with_artist(song, artist):
    with open("songs.txt", "a") as file:
        file.write(f"{song} by {artist}\n")
```

```python
write_song_with_artist("Shape of You", "Ed Sheeran")
write_song_with_artist("Blinding Lights", "The Weeknd")
write_song_with_artist("Rolling in the Deep", "Adele")

with open("songs.txt", "r") as file:
    lines = file.readlines()
    for line in lines:
        print(line, end="")
```
