# Interactive Input Field
This notebook demonstrates how to use ipywidgets to create an interactive input field in Jupyter.

```python
# Install ipywidgets if needed
import subprocess
import sys

try:
    import ipywidgets
except ImportError:
    subprocess.check_call([sys.executable, "-m", "pip", "install", "ipywidgets"])
```

```python
from ipywidgets import Text, Button, Output, VBox
from IPython.display import display

# Create widgets
text_input = Text(
    placeholder='Type something...',
    description='Input:',
    style={'description_width': '50px'}
)

button = Button(
    description='Submit',
    button_style='success'
)

output = Output()

# Define button click behavior
def on_button_click(b):
    with output:
        output.clear_output()
        user_text = text_input.value
        if user_text:
            print(f"✓ You entered: {user_text}")
        else:
            print("⚠ Please enter something!")

button.on_click(on_button_click)

# Display the interface
display(VBox([text_input, button, output]))
```
