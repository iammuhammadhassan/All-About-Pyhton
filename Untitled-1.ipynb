```python
import os
from google import genai
from google.genai import types
from dotenv import load_dotenv
```

```python
load_dotenv()
api_key = os.getenv("GOOGLE_API_KEY") or os.getenv("GEMINI_API_KEY")
client = genai.Client(api_key=api_key) if api_key else None
```

```python
SYSTEM_PROMPT = """
You are a helpful and friendly medical assistant.

Rules:
- Explain health topics in simple language.
- Do NOT provide diagnoses.
- Do NOT prescribe medication or dosages.
- Always encourage consulting a doctor for serious issues.
- If question is risky, respond cautiously.
- Keep tone calm, clear, and supportive.
"""

def is_unsafe_query(user_input):
    unsafe_keywords = [
        "suicide", "kill myself", "overdose",
        "self harm", "dosage", "how many pills"
    ]
    return any(word in user_input.lower() for word in unsafe_keywords)

def get_response(user_input):
    if is_unsafe_query(user_input):
        return "I'm really sorry you're going through something difficult. It's best to talk to a doctor or a trusted person right away."

    if client is None:
        return "Gemini API key not found. Set GOOGLE_API_KEY (or GEMINI_API_KEY) in your .env file."

    response = client.models.generate_content(
        model=os.getenv("GEMINI_MODEL", "gemini-2.5-flash"),
        contents=f"System instruction:\n{SYSTEM_PROMPT}\n\nUser: {user_input}",
        config=types.GenerateContentConfig(temperature=0.5)
    )

    return response.text or "I couldn't generate a response right now."


try:
    import ipywidgets as widgets
    from IPython.display import display
except ImportError:
    widgets = None

print("Health Chatbot")
print("Type your message in the box below and click Send, or press Enter if supported.\n")

if widgets is None:
    print("ipywidgets is not installed in this environment, so notebook input cannot be shown.")
    print("Install ipywidgets or run the chatbot as a Python script to use input().")
else:
    chat_output = widgets.Output(layout={"border": "1px solid #d0d7de", "padding": "10px", "height": "260px", "overflow_y": "auto"})
    user_input = widgets.Text(
        placeholder="Type your health question here",
        description="You:",
        layout=widgets.Layout(width="75%")
    )
    send_button = widgets.Button(description="Send", button_style="primary")

    def handle_send(_=None):
        text = user_input.value.strip()
        if not text:
            return

        reply = get_response(text)
        with chat_output:
            print(f"You: {text}")
            print(f"Bot: {reply}")
            print()
        user_input.value = ""

    send_button.on_click(handle_send)
    if hasattr(user_input, "on_submit"):
        user_input.on_submit(handle_send)

    display(widgets.VBox([chat_output, widgets.HBox([user_input, send_button])]))
```

```python
print("Chatbot is ready. Type your message above and click Send.")
input("Press Enter to exit the chatbot...")
```
