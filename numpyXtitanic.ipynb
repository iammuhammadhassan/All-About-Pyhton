```python
count=0
survived=0
for i in passengers:
    if(i[1] == 1):
        print(i)
        survived+=1
    count+=1
print('Total surviced : ',survived)
print('Average of survival: ',survived/count)
```

Now i will Find Percentage of passengers survived based on their Class

```python
upper_class=[]
middle_class=[]
lower_class=[]
for i in passengers:
    if(i[2] == 1):
        upper_class.append(i)
    elif(i[2] == 2):
        middle_class.append(i)
    else:
        lower_class.append(i)

```

```python
upper_class=np.array(upper_class)
middle_class=np.array(middle_class)
lower_class=np.array(lower_class)
```

```python
upper_class_survived = 0
for i in upper_class:
    if(i[1] == 1):
        upper_class_survived+=1
print('Upper class survived : ',upper_class_survived)
print("Average :",upper_class_survived/len(upper_class))
 
```

```python
middle_class_survived = 0
for i in middle_class:  
    if(i[1] == 1):
        middle_class_survived+=1 
print('Middle class survived : ',middle_class_survived)
print("Average :",middle_class_survived/len(middle_class))
```

```python
lower_class_survived = 0
for i in lower_class:
    if(i[1] == 1):
        lower_class_survived+=1
print('Lower class survived : ',lower_class_survived)
print("Average :",lower_class_survived/len(lower_class))
```
